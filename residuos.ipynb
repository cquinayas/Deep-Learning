```
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Convolution2D, MaxPooling2D, Flatten, Dense
```

```
#Armar modelo
#Inicializar modelo

model = Sequential()

#Agregar la primera Convolución de la red neuronal 
#64 filtros
model.add(Convolution2D(64, (3, 3), strides=(1,1), input_shape=(32,32,3), activation='relu'))#agregando las capas convencionales
```

```
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Convolution2D(32, (3,3), strides=(1,1), activation='relu'))# Añadimos la segunda capa
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
```

```
model.add(Flatten())
```

```
#Conexion completa
model.add(Dense(128, activation='relu'))
#Capa de salida
model.add(Dense(1, activation='sigmoid'))
```

```
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```
# Creamos unas modificaciones sobre las imágenes para tener más cantidad de imágenes con las que entrena
train_data_gen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
)

test_datagen = ImageDataGenerator(
    rescale = 1./255
)
```

```
Entrenamiento_set = train_data_gen.flow_from_directory( 
    "Entrenamiento",
    target_size = (32, 32),
    batch_size = 128, 
    class_mode = "binary"
)

test_set = test_datagen.flow_from_directory("Validacion",
                                            target_size = (32,32),
                                            batch_size = 128,
                                            class_mode = "binary"
                                            )
```

```
model.fit_generator(
    Entrenamiento_set,
    epochs = 30,
    validation_data = test_set,
)
```
